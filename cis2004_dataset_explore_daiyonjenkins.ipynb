```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by 'pH Level' and calculate the mean of 'Species Observed'
ph_species_data = df.groupby('pH Level')['Species Observed'].mean().reset_index()

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='pH Level', y='Species Observed', data=ph_species_data, palette='viridis')
plt.title('Average Species Observed by pH Level')
plt.xlabel('pH Level')
plt.ylabel('Average Number of Species Observed')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
